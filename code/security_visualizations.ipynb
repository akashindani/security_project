```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open("Security_research/100_percent_label_lstm_f1.txt", "rb") as fp:   # Unpickling
    percent_100_label_lstm_f1 = pickle.load(fp)

with open("Security_research/1_percent_label_lstm_f1.txt", "rb") as fp:   # Unpickling
    percent_1_label_lstm_f1 = pickle.load(fp)

with open("Security_research/1_percent_with_gan_lstm_f1.txt", "rb") as fp:   # Unpickling
    percent_1_with_gan_lstm_f1 = pickle.load(fp)
```

```python
# summarize history for loss
plt.plot(percent_100_label_lstm_f1)
plt.plot(percent_1_label_lstm_f1)
plt.plot(percent_1_with_gan_lstm_f1)
plt.title('All models F1 score')
plt.ylabel('F1-score')
plt.xlabel('epoch')
plt.legend(['100 % data + LSTM', '1 % data + LSTM','1 % data + GAN + LSTM'], loc='lower right')
plt.show()
```
