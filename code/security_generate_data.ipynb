```python
'''
def calc_src_ip_distrib(start,end, src_ip):
    ans = df[(df['realtive_time'] < end) & (df ['realtive_time'] >= start) & (df ['src_ip'] == src_ip)]['datetime'].count()/df[(df['realtive_time'] < end) & (df ['realtive_time'] >= start)]['datetime'].count()
    return round(ans, rounding_digits)

'''

def calc_dst_ip_distrib(start,end, dst_ip):
    ans = df[(df['realtive_time'] < end) & (df ['realtive_time'] >= start) 
             & (df ['dst_ip'] == dst_ip)]['datetime'].count()/(df[(df['realtive_time'] < end) 
                                                                 & (df ['realtive_time'] >= start)]['datetime'].count() + 1)
    return round(ans, rounding_digits)


def calc_referer_distrib(start,end, referer):
    if referer == 0:
        return 0.0
    ans = df[(df['realtive_time'] < end) & (df ['realtive_time'] >= start) 
             & (df ['referer'] == referer)]['datetime'].count()/(df[(df['realtive_time'] < end) 
                                                                   & (df ['realtive_time'] >= start) 
                                                                   & (df ['referer'] != 0)]['datetime'].count() + 1)
    return round(ans, rounding_digits)

def calc_user_agent_distrib(start,end, user_agent):
    if user_agent == 0:
        return 0.0
    ans = df[(df['realtive_time'] < end) & (df ['realtive_time'] >= start) 
             & (df ['user_agent'] == user_agent)]['datetime'].count()/(df[(df['realtive_time'] < end) 
                                                                         & (df ['realtive_time'] >= start) 
                                                                         & (df ['user_agent'] != 0)]['datetime'].count() + 1)
    return round(ans, rounding_digits)

def calc_time_gap(values, cur_realtive_time):
    if len(values) == 0:
        return round(cur_realtive_time, rounding_digits),0.0
    else:
        return round(cur_realtive_time, rounding_digits), round(cur_realtive_time - values[-1][3], rounding_digits)
```

```python
def caculate_all_values(start,end, src_ip, dst_ip, referer, user_agent, values, cookie_flag):
    inner_list=[]
    
    '''
    #first - calculate src ip distribution
    ans1 = calc_src_ip_distrib(start,end, src_ip)
    inner_list.append(round(ans1, rounding_digits))
    '''
    
    #first - calculate dst ip distribution
    ans1 = calc_dst_ip_distrib(start,end, dst_ip)
    inner_list.append(ans1)
    
    #second - calculate referer distribution
    ans2 = calc_referer_distrib(start,end, referer)
    inner_list.append(ans2)
    
    #third - calculate user_agent distribution
    ans3 = calc_user_agent_distrib(start,end, user_agent)
    inner_list.append(ans3)
    
    #fourth - calculate time gap
    ans4 = calc_time_gap(values, end)
    inner_list.append(ans4[0])
    inner_list.append(ans4[1])
    
    #fifth - take cookie flag
    inner_list.append(float(cookie_flag))
    
    return inner_list
    
```

```python
def check_bot_label(ip,seq_start_time, seq_end_time):
    if ip in bots_dict:
        time_list=bots_dict[ip]
        for time_frame in time_list:
            #print(time_frame, seq_start_time, seq_end_time)
            #if seq_start_time >= time_frame[0] and seq_end_time<= time_frame[1]:
            if (seq_end_time >= time_frame[0] and seq_end_time<= time_frame[1]) or (seq_start_time >= time_frame[0] and seq_start_time<= time_frame[1]):
                #print("returning 1")
                return 1
    return 0
```

```python
def check_bot_for_each_request(dst_ip, relative_time):
    if dst_ip in bots_dict:
        time_list=bots_dict[dst_ip]
        for time_frame in time_list:
            if relative_time >= time_frame[0] and relative_time <= time_frame[1]:
                return 1
    return 0
```
